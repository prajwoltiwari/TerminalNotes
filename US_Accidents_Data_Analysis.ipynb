### Start Time analysis

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
sns.histplot(df.Start_Time.dt.hour, bins=24)
```

- A high percentage of accidents occur between 11am to 6pm. 

```python
sns.histplot(df.Start_Time.dt.dayofweek, bins=7)
```

- The accidents are lower on weekends and are almost evenly distributed in the weekdays

- Is the distribution of number accidents by hour the same on weekends as on weekdays???


```python
sunday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sunday_start_time.dt.hour)
```

```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour)
```

- On Sundays, the peak occurs between 10am and 8pm, unlike weekdays

```python
sns.histplot(df.Start_Time.dt.month, bins=12)
```

- Need to dig deeper into the number of accidents per month. Seems like there are more accidents in the end of the year when the winter is there. More specifically, highest why in the december?
- Does more accidents occur in the winters?

```python
df_by_2019 = df[df.Start_Time.dt.year == 2019]
sns.distplot(df_by_2019.Start_Time.dt.month, bins = 12)
```

```python
df_by_2016 = df[df.Start_Time.dt.year == 2016]
sns.distplot(df_by_2019.Start_Time.dt.month, bins = 12)
```

Much data might be missing in different years due to which the distribution shows more accidents in Deceber.

- We also might have different sources of data. It might be important to compare the data from different sources.

```python
import folium
from folium.plugins import HeatMap
```

```python
lat, lng = df.Start_Lat[0], df.Start_Lng[0]
lat, lng
```

```python
map = folium.Map()
```

```python
marker = folium.Marker((lat, lng))
marker.add_to(map)
map
```

```python
list(zip(list(df.Start_Lat), list(df.Start_Lng)))
```

```python
sample1_df = df.sample(int(0.01*len(df)))
```
